```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
datadirs = ["uni10", "uni1", "1d10", "1d1"]
ABCD = {"uni10": "(a)", "uni1": "(b)", "1d10": "(c)", "1d1": "(d)"}
steps = [0,100, 500, 5000,10000, 50000]
```

```python
# WGAN with GP
fig = plt.figure()
data_1d = np.loadtxt("1d.csv",  delimiter=",")
data_uni = np.loadtxt("uni.csv",  delimiter=",")
plt.rc('font', family='serif')

i = 0
for datadir in datadirs:
    for step in steps:
        i += 1
        ax = plt.subplot(len(datadirs), len(steps),i)
        if "uni" in datadir:
            data_ref = data_uni
            plt.xlim([-1.1, 1.1])
            plt.ylim([-1.1, 1.1])
            plt.yticks([-1, 0, 1])
            plt.xticks([-1, 0, 1])
        if "1d" in datadir:
            data_ref = data_1d
            plt.xlim([-1.3, 1.3])
            plt.ylim([-1.3, 1.3])
            plt.yticks([-1, 0, 1])
            plt.xticks([-1, 0, 1])

        plt.gca().set_aspect('equal', adjustable='box')
        data = np.loadtxt(datadir + "/" + str(step) + ".csv",  delimiter=",")
        plt.scatter(data_ref[:,0], data_ref[:,1], marker='o', color='k', label='Real Data', s = 2)
        plt.scatter(data[:,0], data[:,1], marker='o', color='r', label='Generated Data', s = 2)
        if "uni10" in datadir:
            if step == 0:
                plt.title("training steps: " + str(step), fontsize = 15)    
            else:
                plt.title(str(step),fontsize = 15)
                
        if step == 0:
            plt.ylabel(ABCD[datadir],rotation=0, fontsize = 15)
plt.subplots_adjust(bottom=0, left = 0, right=0.3 * len(steps), top=0.4 * len(datadirs))
```
